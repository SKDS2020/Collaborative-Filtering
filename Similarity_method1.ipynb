```python

def count_identical_searches(x , user_searches):
    sum=0
    for i,j in user_searches.items():  # i should throw exceptions
        if (x[i]>0):
            sum+=1
    return sum

# calc similarity and drop user's searches from the utility matrix
def calc_similarity_method_1(utility_matrix, user_searches):
    return Finding_similarity.find_similarity(user_searches , count_identical_searches ,utility_matrix).drop(user_searches, axis=1)
```

#### Filtering

#### Filter sessions by at least one city in common

```python
def filter_at_least_one_city_in_common(utility_matrix):
    return utility_matrix[utility_matrix['similarity'] >0] # Removing the sessions without any similar search
```

#### Sort cities by by the amount of their searches

###### For each city, we want to know the sum of the times searched for it along with the cities the user searched for

##### add total row

```python
def add_a_total_row(utility_matrix):
    utility_matrix.loc['total'] = (utility_matrix!=0).sum(axis=0)
    return utility_matrix
```

##### Creating a sorted matrix of cities and the amount of their searches

```python
def get_sorted_city_searches_matrix(utility_matrix):
    return utility_matrix.drop(['similarity'], axis=1).iloc[-1].sort_values(ascending=False) #Removing irrelevant columns and returning a matrix of sorted cities sorted by the amount of their searches

```

##### Return a list of recommendations

```python
def get_top_X_recommendations_list(rec_matrix , topx=0):
    if topx<=0: topx=5 # default
    return rec_matrix.head(topx).index.to_list()
    
```
